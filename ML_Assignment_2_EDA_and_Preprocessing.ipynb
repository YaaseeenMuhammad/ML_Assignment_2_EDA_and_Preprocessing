### Outliers

#### IQR Method

```python
Qtr1_age = df['age'].quantile(0.25)
Qtr3_age = df['age'].quantile(0.75)

IQR = Qtr3_age - Qtr1_age

lower_bound = Qtr1_age - 1.3 * IQR
upper_bound = Qtr3_age + 1.3 * IQR

Age_Outliers = (df['age'] < lower_bound) | (df['age'] > upper_bound)
print("Outliers in 'age' column:\n", df[Age_Outliers])

Age_Non_Outliers = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]
print("\nDataFrame after removing outliers:\n","_" * 75,"\n", Age_Non_Outliers)

print("_" * 75)
print(f'\nNumber of outliers in the age column: {Age_Outliers.sum()}')
```

```python
Qtr1_salary=df['salary'].quantile(0.25)
Qtr3_salary=df['salary'].quantile(0.75)

IQR=Qtr3_salary - Qtr1_salary

LowerBound=Qtr1_salary-1.3*IQR
UpperBound=Qtr3_salary+1.3*IQR

Salary_Outliers=(df['salary'] < LowerBound) | (df['salary'] > UpperBound)
print("outliners",df[Salary_Outliers])

Salary_Non_Outlier=df[(df['salary']>=LowerBound) & (df['salary']<=UpperBound)]
print("\nDataFrame after removing outliers:\n","_" * 75,"\n",Salary_Non_Outlier)

print("_" * 75)
print(f'Number of outliers in salary column:{Salary_Outliers.sum()}')
```
